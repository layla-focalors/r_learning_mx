```r
fab <- c(4,2,1,2,2,3,3,2,1,1)
```

```r
table(fab)
```

```r
pie(table(fab))
```

```r
barplot(table(fab), names=c('봄','여름','가을','겨울'), col=c('pink','skyblue','tomato','white'),main='계절 선호도 조사', xlab='계절',ylab='응답수')
```

```r
pie(table(fab), main='fab season')
```

```r
fab.col <- c(2,3,2,1,1,2,2,1,3,2,1,3,2,1,2)
fab.col
```

```r
barplot(table(fab.col), main='선호하는 색', col=c('white','pink','skyblue'))
```
