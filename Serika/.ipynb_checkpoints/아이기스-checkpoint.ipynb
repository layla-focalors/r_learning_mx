```r
age.A = c(13709, 10974, 7979, 5000, 4250)
age.B = c(17540, 29701, 36209, 33947, 24487)
age.C = c(991, 2195, 5366, 12980, 19007)
```

```r
ds = rbind(age.A, age.B, age.C)
```

```r
ds
```

```r
colnames(ds) = c('1970','1990','2010','2030','2050')
```

```r
ds
```

```r
barplot(ds,
       main = '인구 추정',
       xlab = '연도',
       ylab = '인구수',
       legend = T, col = c('pink','skyblue','purple'))
```

```r
barplot(ds,
       main = '인구 추정',
       xlab = '연도',
       ylab = '인구수',
       legend = T, col = c('pink','skyblue','purple'),
       horiz = T,
       beside = T)
```

```r
pie(ds)
```

```r
barplot(ds,
       main = '인구 추정',
       xlab = '연도',
       ylab = '인구수',
       legend = T, col = c('pink','skyblue','purple'),
       horiz = F,
       beside = T,
       args.legend = list(bty = 'n', x = 'topright', inset = c(-0.25, 0)))
```
