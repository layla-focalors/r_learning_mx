```r
fab.chr = c('레일라','루미네','닐루','라이덴', '라이덴','닐루', '레일라','레일라')
table(fab.chr)
```

```r
pie(table(fab.chr), main='pie chart')
```

```r
pie(table(fab.chr), main='pie chart', radius = 3)
```

```r
pie(table(fab.chr), main='pie chart', radius = 2)
```

```r
pie(table(fab.chr), main='pie chart', radius = 1)
```

```r
install.packages('plotrix')
```

```r
library('plotrix')
```

```r
pie3D(table(fab.chr), labels=names(table(fab.chr)), main='선호하는 캐릭터',
     col = c('pink', 'skyblue', 'white', 'orange'),
     explode=0.1)
```

```r
pie3D(table(fab.chr), labels=names(table(fab.chr)), main='선호하는 캐릭터',
     col = c('pink', 'skyblue', 'white', 'orange'),
     explode=0)
# explode는 파이의 간격!
```

```r
pie3D(table(fab.chr), labels=names(table(fab.chr)), main='선호하는 캐릭터',
     col = c('pink', 'skyblue', 'white', 'orange'),
     explode=0, radius=3)
# explode는 파이의 간격!
```
