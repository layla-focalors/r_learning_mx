```r
library(fmsb)
```

```r
score = c(80, 60, 95, 85, 40)
```

```r
max.score = rep(100, 5)
```

```r
min.score = rep(0, 5)
```

```r
ds = rbind(max.score, min.score, score)
```

```r
ds
```

```r
ds = data.frame(ds)
colnames(ds) = c('지능','힘','마력','논리력','명중률')
```

```r
radarchart(ds)
```

```r
radarchart(ds,
          pcol = 'pink',
          pfcol = rgb(255,211,182, 0.5),
          plwd = 3,
          cglcol = 'pink',
          cglty = 1, 
          cglwd = 0.8,
          axistype = 1,
          seg = 4, 
          axislabcol = 'black',
          caxislabels = seq(0, 100, 25))
```
